```python
import pandas as pd
df = pd.read_csv('movie_sample_dataset.csv')
df.head()
```

Membaca dataset movie_sample_dataset.csv ke dalam object df kemudian menampilkan 5 baris data teratas dengan menggunakan library pandas yang diimport sebagai pd

```python
df.dtypes
```

Memeriksa tipe data pada tiap variabel

```python
import numpy as np
df.replace([" ","Nan", "N/A", "n/a", "null", "NULL", "Null", "-"], np.nan, inplace=True)
print(df.isnull().sum())
```

* Mengidentifikasi nilai yang hilang dengan menggunakan library numpy dan diimport sebagai np
* Mengganti nilai kosong, "N/A", "Null", dan "-" dengan nilai "nan" kemudian menampilkan banyakknya data yang memiliki nilai "nan" sehingga teridentifikasi sebagai nilai yang hilang pada setiap kolom variabel



```python
df.dropna(subset=["color", "director_name", "gross", "genres", "budget"], axis=0, inplace=True)
df.reset_index(drop=True, inplace=True)
```

Menghapus nilai "nan" yang telah teridentifikasi sebagai nilai yang hilang tersebut

```python
df=df[(df["duration"] >= 30) & (df["duration"] < 300)]
df=df[(df["imdb_score"] >= 0) & (df["imdb_score"] <= 10)]
```

Menangani Outliner data yang tidak masuk akal pada variabel "duration", "imdb_score"

```python
df.drop_duplicates(inplace=True)
```

Menghapus baris data yang terduplikasi

```python
df["color"]=df["color"].str.capitalize()
df["country"]=df["country"].str.upper()
df["language"]=df["language"].str.capitalize()

df["country"]=df["country"].replace({"UNITED STATES":"USA", "NEW ZEALAND":"NZ", "FRANCE":"FR", "AUSTRALIA":"AU", "CZECH REPUBLIC":"CZ"})
```

Menormalisasi data yang tidak konsisten seperti perbedaan ("Color" dan "color"), ("USA", "usa", dan "United States")  pada variabel "color", "country", dan "language"

```python
df.to_csv('movie_dataset_cleaned.csv', index=False)
```
